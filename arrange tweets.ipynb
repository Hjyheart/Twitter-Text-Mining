```python
from elasticsearch import Elasticsearch
import pandas as pd
import numpy as np
import os
```

```python
es = Elasticsearch()
```

```python
democracy = pd.read_csv('./democracy.csv')
```

```python
republican = pd.read_csv('./republican.csv')
```

```python
def get_query_body(id_str):
    return {
      "query": {
        "match": {
            "u_id": id_str
        }
      }
    }
```

```python
res = es.search(index='tweets', body=get_query_body('904662446'), size=50)
```

```python
res['hits']['hits'][0]['_source']['full_text']
```

```python
index = 0
for user in republican.iterrows():
    user = user[1]
    print('Deal with user: {} {}'.format(user['id'], index))
    index += 1
    try:
        res = es.search(index='tweets', body=get_query_body(user['id']), size=50)
        for tweet in res['hits']['hits']:
            with open('./tweets_r_all/{}/{}_{}.txt'.format(index%25, user['id'], tweet['_source']['id']), 'w') as f:
                f.write(tweet['_source']['full_text'])
                f.close()
    except:
        print("Fetch user: {} fail".format(user['id']))
```

```python
dfs = []
for item in os.walk('./tweets_d_all_csv/'):
    file_names = item[2]
    for file_name in file_names:
        dfs.append(pd.read_csv('./tweets_d_all_csv/' + file_name))
```

```python
pd.concat(dfs).to_csv('democracy_tweets.csv')
```
